```
import pandas as pd
import plotly.graph_objects as go

df1 = pd.read_csv('./data/refrigDoubleBox.csv')
df2 = pd.read_csv('./data/refrigSingleBox.csv')

fig1 = go.Figure(go.Scatter(x = df1['Time'], y = df1['Temperature'],
                  name='Double Box'))

fig1.update_layout(title='Refrigerator Test Result _ Temperature',
                   plot_bgcolor='rgb(230, 230,230)',
                   showlegend=True)

fig2 = go.Figure(go.Scatter(x = df2['Time'], y = df2['Temperature'],
                  name='Single Box'))

fig2.update_layout(title='Refrigerator Test Result _ Temperature',
                   plot_bgcolor='rgb(230, 230,230)',
                   showlegend=True)

fig1.show()
fig2.show()
```

```
import pandas as pd
import numpy as np

DBTemp_filepath = './data/refrigDoubleBox.csv' 
SBTemp_filepath = './data/refrigSingleBox.csv' 

DBTemp = pd.read_csv(DBTemp_filepath, header = 0, index_col = 0,
                         parse_dates=True, infer_datetime_format=True)
SBTemp = pd.read_csv(SBTemp_filepath, header = 0, index_col = 0,
                       parse_dates=True, infer_datetime_format=True)


ax = DBTemp.plot(figsize=(20,10))
SBTemp.plot(ax=ax)
```
